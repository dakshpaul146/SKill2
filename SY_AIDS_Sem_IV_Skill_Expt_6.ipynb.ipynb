* **Scatter Plot**: sns.scatterplot with hue to differentiate points
  by the 'origin' category.
* **Histogram**: sns.histplot for a distribution plot of 'mpg' with a kernel density estimate (KDE).
* **Bar Plot**: sns.barplot to show the mean of 'mpg' by 'origin'.
* **Line Chart**: sns.lineplot for a line chart of 'mpg' over 'model_year' with
  markers.
* **Box Plot**: sns.boxplot to visualize the distribution of 'mpg' across different 'origin' categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
```

* **Scatter plot**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x= df['mpg'], y=df['hp'], palette='viridis', data=df)
plt.title('Scatter Plot with Hue')
plt.show()
```

* **Histogram**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['mpg'], bins=10, kde=True)
plt.title('Histogram of MPG')
plt.show()
```

* **Bar Plot**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df['drat'], y=df['mpg'], data=df)
plt.title('Bar Plot')
plt.show()
```

* **Line Chart**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['carb'], y=df['mpg'], data=df, marker='o')
plt.title('Line Chart')
plt.show()
```

* **Box Plot**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['drat'], y=df['mpg'], data=df)
plt.title('Box Plot')
plt.show()
```
